```python
from pylab import plot, show

y=[1.,2.4,1.7,0.3,0.6,1.8]

plot(y)
show()
```

```python
from numpy import linspace,sin,cos

x=linspace(0,10,1000)
y=sin(x)

plot(x,y)
show()
```

```python
from pylab import ylim,xlim,grid

x=linspace(0,10,1000)
y=sin(x)

plot(x,y)
ylim=(-1.1,1.1)
xlim=(0,10)
grid("on")
show()
```

```python
from pylab import ylabel,xlabel

x=linspace(0,10,1000)
y=sin(x)

plot(x,y)
ylim=(-1.1,1.1)
xlim=(0,10)
grid("on")

ylabel("Eje y")
xlabel("eje x")

show()
```
